Tip:

2. Utiliza la función `len()`

````python
# Ejecuta esta celda

mylist = [1, 2, 3, 4, 5, 6, 4, 3, 2, 1, 2, 3, 4, 5, 6, 6, 7, 8, 5, 6, 7, 8,
          9, 8, 9, 8, 9, 7, 10, 1, 23 ,1 ,2 ,2 ,3 ,1 ,3 ,2 ,4 ,1 ,4 ,4 ,1 ,
          2, 22, 3, 4, 1, 4, 1, 3, 1, 22, 3, 6, 3, 1, 23, 4, 6, 1, 2, 10, 1]
````

````python
# Tu código aquí ...
print(
    f'La cantidad de enteros únicos en la lista mylist es: '
    f'{sum([isinstance(x,int) for x in set(mylist)])}.'
)
# Se ajusto el el uso de len por sum, para verificar si los elementos del
# conjunto son enteros, y hacer solo la cuenta sobre estos.
# Si omitimos dicho ajuste el comando debería quedar como len(set(mylist)).
````
