## <font color='green'>Tarea 1: Challenging</font>
Cree una clase Esfera que acepte un radio en la instanciación y tenga  métodos de volumen y área de superficie.

Tip:

* Utiliza la función `round( )` para ajustar los decimales de salida

```python
# Tu código aquí ...

class Esfera():

  def __init__(self,radio:float = 1):
    self.radio = radio

  def __str__(self):
    return f"Esfera de radio {self.radio}."

  def area_superficie(self):
    pi = 3.141592653589793
    return 4*pi*self.radio**2

  def volumen(self):
    pi = 3.141592653589793
    return (4*pi*self.radio**3)/3
```

```python
s = Esfera(1)
```

```python
s.area_superficie()
```

```python
s.volumen()
```

4. Crea un método en la clase llamado *reset_random* que vuelva a generar un número aleatorio (también entre 0-10) para el atributo `self.rand_choice`
5. Crea un método llamado *guess* (adivinar) que use la función `input()` para aceptar un número del usuario. Este método debe imprimir una declaración que indique si la suposición del usuario fue correcta o no.

Bonus track: agrega una lógica en el método *guess* que informe al usuario cuando su número está por arriba o por debajo del número a adivinar.

Tip:
* En el punto 3 use `import random`
* Te dejamos el esqueleto de la clase

```python
# Tu código aquí ...

# Esqueleto de la clase
# class GuessingGame():
#     def __init__(self):
#       pass

#     def reset_random(self):
#         pass

#     def guess(self):
#         pass

# Solución
class GuessingGame():
    def __init__(self):
        from random import randint
        self.rand_choice = randint(0,10)

    def reset_random(self):
        from random import randint
        self.rand_choice = randint(0,10)

    def guess(self):
        num = int(input("Ingrese un número entero entre 1 y 10:"))
        if num == self.rand_choice:
          return f"Acertaste en el número! Felicidades!\nLLama al método reset_random() para seguir jugando."
        elif num < self.rand_choice:
          return print(f"Apuesta más alto!\nLLama al método guess() para intentar nuevamente.")
        else:
          return print(f"Apuesta más bajo!\nLLama al método guess() para intentar nuevamente.")
```

```python
g = GuessingGame()
```

```python
g.reset_random()
```

```python
g.guess()
```

### Modelando un zoológico
Crea una jerarquía de clases para modelar un zoológico. Debe haber una clase Animal con atributos comunes a todos los animales. Luego, crea clases para diferentes tipos de animales (por ejemplo, Mamífero, Reptil, Ave), cada una de las cuales hereda de Animal. Cada una de estas clases debe tener atributos y métodos específicos de su tipo. Por último, crea clases para animales específicos (por ejemplo, León, Cocodrilo, Águila) que heredan de sus respectivas clases de tipo de animal. Utiliza el polimorfismo para crear un método en la clase Animal que describa el sonido que hace el animal, pero implementa este método de manera diferente en cada clase de animal específico.

```python
# Tu código aquí ...
class Animal():

  def __init__(self, nombre_comun, nombre_cientifico, sonido):
    self.nombre_comun = nombre_comun
    self.nombre_cientifico = nombre_cientifico
    self.sonido = sonido

  def __str__(self):
    return f"{self.nombre_cientifico}, animal de nombre común {self.nombre_comun}."

  def desc_sonido(self):
    return print(f"La vocalizaón de {self.nombre_cientifico} está dado por la onomatopeya {self.sonido}")

class Mamifero(Animal):

  def __init__(self, nombre_comun, nombre_cientifico, sonido, n_camada, pelaje):
    super().__init__(nombre_comun, nombre_cientifico, sonido)
    self.n_camada = n_camada
    self.pelaje = pelaje

  def __str__(self):
    return f"Mamífero de nombre científico {self.nombre_cientifico}, comúnmente llamdo {self.nombre_comun}"

  def desc_sonido(self):
    return f"Este mamífero vocaliza con la onomatopeya {self.sonido}"

class Perro(Mamifero):

  def __init__(self, nombre_cientifico, nombre_comun, sonido, n_camada, pelaje, raza, dueno, nombre):
    super().__init__(nombre_cientifico, nombre_comun, sonido, n_camada, pelaje)
    self.raza = raza
    self.dueno = dueno
    self.nombre = nombre

  def __str__(self):
    return f"Perro llamado {self.nombre} de la raza {self.raza}, mascota de {self.dueno}"

  def desc_sonido(self):
    return f"{self.nombre} ladra {self.sonido}."
```

```python
perro = Animal('Perro', 'Canis lupus', 'woof')
perro.desc_sonido()
```

```python
perro2 = Mamifero("perro", "Canis lupus", "woof", "de 2 a 10", "variado")
perro.desc_sonido()
```

```python
perro3 = Perro("perro", "Canis lupus", "woof", "de 2 a 10", "café", "corgi", "Ed", "Ein")
perro3.desc_sonido()
```

### Modelando una biblioteca
Crea una jerarquía de clases para modelar una biblioteca. Debe haber una clase Libro con atributos comunes a todos los libros. Luego, crea clases para diferentes tipos de libros (por ejemplo, Novela, CienciaFiccion, Biografia), cada una de las cuales hereda de Libro. Cada una de estas clases debe tener atributos y métodos específicos de su tipo. Utiliza el polimorfismo para crear un método en la clase Libro que describa el tipo de lectura que proporciona el libro, pero implementa este método de manera diferente en cada clase de libro específico.

```python
# Tu código aqui ...

class Libro():
  def __init__(
              self,
              titulo: str,
              autor: str,
              editorial: str,
              paginas: int,
              codigo: str,
              topicos: list[str],
              resumen: str,
              anio: int
      ):
    self.titulo = titulo
    self.autor = autor
    self.editorial = editorial
    self.paginas = paginas
    self.codigo = codigo
    self.topicos = topicos
    self.resumen = resumen
    self.anio = anio

  def __str__(self):
    return f"Libro titulado {self.titulo} escrito por {self.autor}, publicado por {self.editorial} el año {self.anio}."

  def descripcion(self):
    print(f"Resumen de {self.titulo}:")
    print(f"{self.resumen}")
    return self.topicos

  def topics(self):
    print(f"El libro {self.titulo} está categorizado en los siguientes tópicos:")
    for topic in self.topicos:
      print(topic)
    return self.topicos

class Novela(Libro):

  def __init__(
              self,
              titulo: str,
              autor: str,
              editorial: str,
              paginas: int,
              codigo: str,
              topicos: list[str],
              resumen: str,
              anio: int,
              genero: str,
              ):
    super().__init__(titulo,autor,editorial,paginas,codigo, topicos, resumen, anio)
    self.genero = genero

  def descripcion(self):
    super().descripcion()
    print(f"Novela del genero {self.genero}.")

class CienciaFiccion(Libro):

  def __init__(
              self,
              titulo: str,
              autor: str,
              editorial: str,
              paginas: int,
              codigo: str,
              topicos: list[str],
              resumen: str,
              anio: int,
              tiempo: str
              ):
    super().__init__(titulo,autor,editorial,paginas,codigo, topicos, resumen, anio)
    self.tiempo = tiempo

  def descripcion(self):
    super().descripcion()
    print(f"Ambientada en {self.tiempo}.")

class Biografia(Libro):

  def __init__(
              self,
              titulo: str,
              autor: str,
              editorial: str,
              paginas: int,
              codigo: str,
              topicos: list[str],
              resumen: str,
              anio: int,
              personaje_principal: str,
              resumen_personaje: str,
              ):
    super().__init__(titulo,autor,editorial,paginas,codigo, topicos, resumen, anio)
    self.personaje_principal = personaje_principal
    self.resumen_personaje = resumen_personaje

  def descripcion(self):
    super().descripcion()
    print(f"Basado en la vida de {self.personaje_principal}, {self.resumen_personaje}.")
```
