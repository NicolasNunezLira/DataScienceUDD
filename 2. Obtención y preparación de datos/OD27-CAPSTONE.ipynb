Basado en la estructura del objeto **iris_obj**, convierta este en un dataframe.

### <font color='green'>Actividad 2</font>

1. Reemplace los valores de la columna "species" por los siguientes valores:
  * "setosa" = 0
  * "versicolor" = 1
  * "virginica" = 2

2. Agrupe las muestras por especies
3. Sin usar el método `describe` calcule los estadísticos básicos de la muestra:
  * cantidad de elementos no nulos
  * media
  * varianza
  * Desviación estándar
  * Percentiles y cuartiles
  * Rango inter-cuartílico
  * Valores máximos y mínimos
  * Rango (diferencia entre el máximo y el mínimo)
4. Repita algunos de los análisis anteriores, pero muestre sus resultados por columnas y especies.
5. Compare sus resultados usando el metodo `describe` por columnas y por especies
6. Generar una tabla multi-indice por "especies" y "ancho del pétalo" y que de cuenta de la suma de los valores de cada clase



````python
"""
1.  Reemplace los valores de la columna "species" por los siguientes valores:

    -"setosa" = 0
    -"versicolor" = 1
    -"virginica" = 2
"""
df['species'] =df['species'].map(
    lambda x: "setosa" if x==0 else (
        "versicolor" if x==1 else "virginica"
    ))
display(df)
````

````python
"""
2.  Agrupe las muestras por especies
"""
df.groupby(by='species').mean()
````

````python
"""
3.  Sin usar el método describe calcule los estadísticos básicos de la muestra
"""
# Cantidad de elementos no nulos
def nnz(x):
    return sum(x!=0)

# Promedio
def media(x):
    return sum(x)/len(x)

# Varianza
def varianza(x):
    return sum([(i-media(x))**2 for i in x])/len(x)

# Percentiles
def q1(x):
    return x.quantile(.25)

def q2(x):
    return x.quantile(.5)

def q3(x):
    return x.quantile(.75)

# Rango intercuartilico
def iqr(x):
    return q3(x)-q1(x)

# Rango
def rango(x):
    return max(x)-min(x)

display(round(
    df.iloc[:,0:4].apply((nnz, media, varianza,
    q1, q2, q3, iqr, 'min', 'max', rango)),2))
````

````python
"""
4.  Repita algunos de los análisis anteriores, pero muestre sus resultados por
    columnas y especies.
"""
display(round(
    df.groupby('species').agg((nnz, media, varianza,
    q1, q2, q3, iqr, 'min', 'max', rango)),2)
)
````

````python
"""
5.  Compare sus resultados usando el metodo describe por columnas y por
    especies.
"""
display(round(
    df.groupby(by='species').describe(),2))
````

````python
"""
6.  Generar una tabla multi-indice por "especies" y "ancho del pétalo" y que de
    cuenta de la suma de los valores de cada clase
"""
df.groupby(by=['species','petal width (cm)']).sum()
````

````python
import seaborn as sns
titanic = sns.load_dataset("titanic")
titanic = titanic[['survived', 'pclass', 'sex', 'age', 'fare', 'embarked']]
titanic.head(5)
````

````python
# Imputación de datos faltantes con el promedio de la edad
titanic.fillna({'age': titanic['age'].mean(skipna=True)}, inplace=True)

# Eliminación de filas con datos faltantes luego de la imputación
titanic.dropna(axis=0, how='any', inplace=True)
````

2. Genere una tabla que muestre la cantidad de personas agrupadas por sexo y clase.
3. Muestre la misma información del punto 1, pero la cantidad de sobrevivientes.



````python
"""
1.  Entregar la información correspondiente a la la tasa de supervivencia
    promedio agrupada por sexo y clase del pasajero.
"""
display(round(
    titanic.groupby(by=['sex', 'pclass'])['survived'].mean(),2))
````

Hasta ahora hemos examinado las tasas de supervivencia según el sexo y la clase de pasajeros. Otro factor que podría haber influido en la supervivencia es la edad. **¿Los niños varones tenían tanta probabilidad de morir como las mujeres en general?** Podemos investigar esta pregunta mediante la indexación múltiple, o pivotando en más de dos variables, agregando otro índice.

Divida la información de la columna "age" en intervalos, creando así una nueva categoría que se puede utilizar para agrupar. Divida el conjunto en intervalos de (0, 12], (12, 18] y (18, 80].

4. Genere la tabla que le ayude a responder la pregunta anterior (considerando los rangos definidos).
5. Para una mejor interpretación de la información anterior, genere una tabla con el número de entradas por cada grupo.






````python
"""
4.  Genere la tabla que le ayude a responder la pregunta anterior
    (considerando los rangos definidos).
"""

# Mapeo de edad a rangos
titanic['age'] = titanic['age'].map(
    lambda x: '(0,12]' if x <= 12 else (
        '(12,18]' if x <= 18 else '(18,80]'
    )
)
````

````python
display(round(
    titanic.groupby(by=['sex','age'])['survived'].mean(), 2))

print(
    f"Podemos ver que los niños varones en el rango de edad (0,12], tenían "
    "considerablemente más posibilidades de morir que las mujeres en general."
)
````

````python
"""
5.  Para una mejor interpretación de la información anterior, genere una tabla
    con el número de entradas por cada grupo.
"""

display(round(
    titanic.groupby(by=['sex','age'])['survived'].agg(('mean', 'count')),2)
)
````

Pandas también admite la indexación múltiple en las columnas. Como ejemplo, considere el precio de los boletos de un pasajero. Esta es otra característica continua que se puede discretizar. Divida los precios en 2 cuantiles iguales (baratos y caros). Algunos de los grupos resultantes podrían estar vacíos; para mejorar la legibilidad de la tabla use un guión en las casillas vacías.

6. Genere una tabla que nos permita ver el efecto del precio de los tickets.

````python
titanic['fare'] = titanic['fare'].map(
    lambda x: 'cheap' if x <= titanic['fare'].quantile(.5) else 'expensive'
)
````

````python
display(round(
    titanic.groupby(by=['sex', 'fare','age'])['survived'].agg(('mean','count')),
    2)
)
````
