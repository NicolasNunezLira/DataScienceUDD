Crear 3 datasets con respuestas de clientes a una encuesta de satisfacción de producto. Las encuestas tienen una escala de 1 a 7 (discreta), donde nota 1 es muy baja satisfacción y la nota 7 es muy alta satisfacción con el producto.
Los datasets son los siguientes:

**Desarrollo**: Guiados por las representaciones gráficas, las cuales es posible interpretar como variables aleatorias distribuidas normalmente, se modelaran las calificaciones a partir de la distribución `scipy.stats.truncnorm`, la cual nos permite ajustar el rango de las calificaciones que distribuidas normalmente.

```python
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Distribución capel
mu1 = 4.5
sigma1 = 1.5
a = (1 - mu1)/sigma1
b = (7 - mu1)/sigma1
norm1 = stats.truncnorm(a, b, loc=mu1, scale=sigma1)
x1=np.linspace(1, 7)
plt.plot(x1, norm1.pdf(x1), 'r')
plt.axvline(mu1, ds='steps', c='b', ls='--', alpha=.5, label=f'Media {mu1}')
plt.grid(alpha=.4, axis='x')
plt.xlabel('Nivel de satisfacción')
plt.title('Distribución calificaciones Capel')
plt.legend()
plt.show()

# Creación dataset
ds_capel = norm1.rvs(800)
```

```python
# Distribución competencia
mu2 = 6.1
sigma2 = 1
a = (3 - mu2)/sigma2
b = (7 - mu2)/sigma2
norm2 = stats.truncnorm(a, b, loc=mu2, scale=sigma2)
x2=np.linspace(3, 7)
plt.plot(x2, norm2.pdf(x2), 'r')
plt.axvline(mu2, ds='steps', c='b', ls='--', alpha=.5, label=f'Media {mu2}')
plt.grid(alpha=.4, axis='x')
plt.xlabel('Nivel de satisfacción')
plt.title('Distribución calificaciones principal competencia')
plt.legend()
plt.show()

# Creación dataset
ds_comp = norm2.rvs(500)
```

```python
# Distribución lepac
mu3 = 6.15
sigma3 = .9
a = (3.5 - mu3)/sigma3
b = (7 - mu3)/sigma3
norm3 = stats.truncnorm(a, b, loc=mu3, scale=sigma3)
x3=np.linspace(3.5, 7)
plt.plot(x3, norm3.pdf(x3), 'r')
plt.axvline(mu3, ds='steps', c='b', ls='--', alpha=.5, label=f'Media {mu3}')
plt.grid(alpha=.4, axis='x')
plt.xlabel('Nivel de satisfacción')
plt.title('Distribución calificaciones Le Pac')
plt.legend()
plt.show()

# Creación dataset
ds_lepac = norm3.rvs(320)
```

## <font color='green'>**Actividad 2**</font>
### Estudiando los datasets

Para cada uno de los datasets creados se pide:

1. Graficar el correspondiente histograma y su función de masa de probabilidad y/o distribución de probabilidad según corresponda.

2. Calcular la media y desviación estandar de la muestra.

3. Calcular el porcentaje de notas 6 y 7. Este indicador en Marketing es conocido como el **Top 2 Box Score**, que corresponde a una proproción de las personas que han manifestado satisfacción por el producto/servicio.

4. Calcular el intervalo de confianza para la media y proporción Top2Box. Ambas con un nivel de significancia alpha del 0.05.




```python
# Datos capel
fig, axs = plt.subplots(1, 2, figsize=(15, 5))
fig.suptitle('Data set Capel')
sns.histplot(ds_capel, ax=axs[0], bins=30, kde=True)
axs[0].set_xlabel('Nivel de satisfacción')
axs[0].set_title('Histograma')
sns.lineplot(x=x1, y=norm1.pdf(x1), ax=axs[1], c='r')
axs[1].set_xlabel('Nivel de satisfacción')
axs[1].set_title('Distribución')
plt.grid(alpha=.4)
plt.show()

# Estadísticos

# Media y desviación estándar
mu = np.mean(ds_capel)
s = np.std(ds_capel)
print(
    f"2. La media y desviación estándar de los datos de Capel es "
    f"{mu:.2f} y {s:.2f}, resp."
)

# Top2Box
t2b = sum(ds_capel>=6)/len(ds_capel)
print(
    f"3. El Top 2 Box Score es {t2b}"
)

# Intevalos de confianza
# Media

s = sigma1/np.sqrt(len(ds_capel))
LI, LS = stats.truncnorm((1-mu)/s, (7-mu)/s, loc=mu, scale=s).interval(.95)
print(
    f"4. El intervalo de confianza para la media muestral para el dataset "
    f"Capel con una significancia de 95% es [{LI:.2f}, {LS:.2f}]."
)

# Proporción
sp = np.sqrt(t2b*(1-t2b)/len(ds_capel))
LI, LS = stats.norm(loc=t2b, scale=sp).interval(.95)
print(
    f"   El intervalo de confianza para la proporción Top2Box para el dataset "
    f"Capel con una significancia de 95% es [{LI:.2f}, {LS:.2f}]."
)
```

```python
# Datos Competencia
fig, axs = plt.subplots(1, 2, figsize=(15, 5))
fig.suptitle('Data set principal competencia')
sns.histplot(ds_comp, ax=axs[0], bins=30, kde=True)
axs[0].set_xlabel('Nivel de satisfacción')
axs[0].set_title('Histograma')
sns.lineplot(x=x2, y=norm2.pdf(x2), ax=axs[1], c='r')
axs[1].set_xlabel('Nivel de satisfacción')
axs[1].set_title('Distribución')
plt.grid(alpha=.4)
plt.show()

# Estadísticos

# Media y desviación estándar
mu = np.mean(ds_comp)
s = np.std(ds_comp)
print(
    f"2. La media y desviación estándar muestral de los datos de la principal competencia son "
    f"{mu:.2f} y {s:.2f}, resp."
)

# Top2Box
t2b = sum(ds_comp>=6)/len(ds_comp)
print(
    f"3. El Top 2 Box Score es {t2b}"
)

# Intevalos de confianza
# Media

s = sigma2/np.sqrt(len(ds_comp))
LI, LS = stats.truncnorm((3-mu)/s, (7-mu)/s, loc=mu, scale=s).interval(.95)
print(
    f"4. El intervalo de confianza para la media muestral para el dataset "
    f"Capel con una significancia de 95% es [{LI:.2f}, {LS:.2f}]."
)

# Proporción
sp = np.sqrt(t2b*(1-t2b)/len(ds_comp))
LI, LS = stats.norm(loc=t2b, scale=sp).interval(.95)
print(
    f"   El intervalo de confianza para la proporción Top2Box para el dataset "
    f"de la principal competencia con una significancia de 95% es [{LI:.2f}, {LS:.2f}]."
)
```

```python
# Datos Competencia
fig, axs = plt.subplots(1, 2, figsize=(15, 5))
fig.suptitle('Data set LePac')
sns.histplot(ds_lepac, ax=axs[0], bins=30, kde=True)
axs[0].set_xlabel('Nivel de satisfacción')
axs[0].set_title('Histograma')
sns.lineplot(x=x3, y=norm3.pdf(x3), ax=axs[1], c='r')
axs[1].set_xlabel('Nivel de satisfacción')
axs[1].set_title('Distribución')
plt.grid(alpha=.4)
plt.show()

# Estadísticos

# Media y desviación estándar
mu = np.mean(ds_lepac)
s = np.std(ds_lepac)
print(
    f"2. La media y desviación estándar muestral de los datos de la principal competencia son "
    f"{mu:.2f} y {s:.2f}, resp."
)

# Top2Box
t2b = sum(ds_lepac>=6)/len(ds_lepac)
print(
    f"3. El Top 2 Box Score es {t2b}"
)

# Intevalos de confianza
# Media

s = sigma2/np.sqrt(len(ds_lepac))
LI, LS = stats.truncnorm((3-mu)/s, (7-mu)/s, loc=mu, scale=s).interval(.95)
print(
    f"4. El intervalo de confianza para la media muestral para el dataset "
    f"LePac con una significancia de 95% es [{LI:.2f}, {LS:.2f}]."
)

# Proporción
sp = np.sqrt(t2b*(1-t2b)/len(ds_lepac))
LI, LS = stats.norm(loc=t2b, scale=sp).interval(.95)
print(
    f"   El intervalo de confianza para la proporción Top2Box para el dataset "
    f"Lepac con una significancia de 95% es [{LI:.2f}, {LS:.2f}]."
)
```

### Creando nuestro test de hipótesis

Desarrollar un test de hipótesis de 1 muestra para al menos 1 de los 3 datasets. Aquí Ud. puede formular la hipótesis a probar. Por ejemplo, puede ser contra un target definido por la industria o retail.



**Solución**:
Para la prueba de hipótesis de una cola, supongamos que la media del nivel de satisfacción de la industria del pisco es de un 4.7, y veamos si la media muestral de Capel es en efecto menor que la media de la industria con una significancia de 95%, esto es:
- $H_0$ : $\mu_{Capel} \leq \mu_{Indutria}$.
- $H_1$ : $\mu_{Capel} > \mu_{Industria}$.

```python
# Test t de una muestra
t_stat, p_value = stats.ttest_1samp(ds_capel, popmean=4.55)

# Ajustar el valor p para un test unilateral (¿media < media_nacional?)
p_value_unilateral = p_value / 2 if t_stat < 0 else 1

# Mostrar resultados
print(f"Estadístico t: {t_stat:.2f}, Valor p (unilateral): {p_value_unilateral:.4e}")

# Nivel de significancia
alpha = 0.05

# Interpretación del resultado
if p_value_unilateral < alpha:
    print("Rechazamos H0: La media muestral es significativamente menor que la media de la industria.")
else:
    print("No podemos rechazar H0: No hay evidencia suficiente para afirmar que la media muestral es menor que la media de la industria.")
```

## <font color='green'>**Actividad 4**</font>
### Validación de algunos test de hipótesis

Desarrollar los siguientes test de hipótesis de 2 muestras para:

1. Verificar si la satisfacción de Le Pac es mayor a la del producto Capel.

2. Verificar si la satisfacción de Le Pac es mayor a la Competencia.

**Soluciones**:
1. Sea $\mu_{LePac}$ y $\mu_{Capel}$ las medias muestrales de Le Pac y Capel respectivamente, y consideremos el test de hipótesis

    - $H_0$ : $\mu_{LePac} \leq \mu_{Capel}$.
    - $H_1$ : $\mu_{LePac} > \mu_{Capel}$.

con una significancia de 95%, esto es posible calcularlo como sigue:

```python
# Nivel de significancia
alpha = 0.05

# Test 1: Le Pac vs. Capel
t_stat, p_value = stats.ttest_ind(ds_lepac, ds_capel, equal_var=False)
p_value_unilateral = p_value / 2 if t_stat > 0 else 1  # Ajustar para una cola

print("Test Le Pac vs. Capel")
print(f"t-stat: {t_stat:.2f}, p-value (unilateral): {p_value_unilateral:.4e}")
if p_value_unilateral < alpha:
    print("Rechazamos H0: La satisfacción de Le Pac es mayor que la de Capel.\n")
else:
    print("No podemos rechazar H0: No hay evidencia suficiente para afirmar que Le Pac es mejor.\n")
```

2. Sea $\mu_{LePac}$ y $\mu_{Comp}$ las medias muestrales de Le Pac y la competencia respectivamente, y consideremos el test de hipótesis

    - $H_0$ : $\mu_{LePac} \leq \mu_{Comp}$.
    - $H_1$ : $\mu_{LePac} > \mu_{Comp}$.

con una significancia de 95%, esto es posible calcularlo como sigue:

```python
# Nivel de significancia
alpha = 0.05

# Test 1: Le Pac vs. Competencia
t_stat, p_value = stats.ttest_ind(ds_lepac, ds_comp, equal_var=False)
p_value_unilateral = p_value / 2 if t_stat > 0 else 1  # Ajustar para una cola

print("Test Le Pac vs. Capel")
print(f"t-stat: {t_stat:.2f}, p-value (unilateral): {p_value_unilateral:.4e}")
if p_value_unilateral < alpha:
    print("Rechazamos H0: La satisfacción de Le Pac es mayor que la de la competencia.\n")
else:
    print("No podemos rechazar H0: No hay evidencia suficiente para afirmar que Le Pac es mejor.\n")
```
